## Exercise: Managing Customer Orders

As part of a business venture, you are starting an online store that sells various products. To ensure smooth operations, you need to develop a program that manages customer orders and inventory.

Follow the steps below to complete the exercise:

1. Define a list called `products` that contains the following items: "t-shirt", "mug", "hat", "book", "keychain".

2. Create an empty dictionary called `inventory`.

3. Ask the user to input the quantity of each product available in the inventory. Use the product names from the `products` list as keys in the `inventory` dictionary and assign the respective quantities as values.

4. Create an empty set called `customer_orders`.

5. Ask the user to input the name of three products that a customer wants to order (from those in the products list, meaning three products out of "t-shirt", "mug", "hat", "book" or "keychain". Add each product name to the `customer_orders` set.

6. Print the products in the `customer_orders` set.

7. Calculate the following order statistics:
   - Total Products Ordered: The total number of products in the `customer_orders` set.
   - Percentage of Products Ordered: The percentage of products ordered compared to the total available products.
   
   Store these statistics in a tuple called `order_status`.

8. Print the order statistics using the following format:
   ```
   Order Statistics:
   Total Products Ordered: <total_products_ordered>
   Percentage of Products Ordered: <percentage_ordered>% 
   ```

9. Update the inventory by subtracting 1 from the quantity of each product. Modify the `inventory` dictionary accordingly.

10. Print the updated inventory, displaying the quantity of each product on separate lines.

Solve the exercise by implementing the steps using the Python concepts of lists, dictionaries, sets, and basic input/output operations. 

### Solution

````python
# Create the data, take input from customer and store it as a dictionary 
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = {}
input_tsh = int(input("Please enter the quantity for the 't-shirt': "))
input_mug = int(input("Please enter the quantity for the 'mug': "))
input_hat = int(input("Please enetr the quantity for the 'hat': "))
input_book = int(input("Please enter the quantity for the 'book': "))
input_keych = int(input("Please enter the quantity for the 'keychain': "))

inventory['t-shirt'] = input_tsh
inventory['mug'] = input_mug
inventory['hat'] = input_hat
inventory['book'] = input_book
inventory['keychain'] = input_keych

print()
# Display dictionary 'inventory' value
print("The data in inventory is: ", inventory)

print()
print()
# Creating a set and store the value from the customer
customer_orders = set()

print("Please choose three products from the following items:", products)
input_1 = input("Item one: ")
input_2 = input("Item two: ")
input_3 = input("item three: ")


# Add the customer input to the 'customer_orders' set
customer_orders.add(input_1)
customer_orders.add(input_2)
customer_orders.add(input_3)

print()
# Display set 'customer_orders' value
print("The products ordered by the customer is: ", customer_orders)

# Calculate customer order statstics
Total_Products_Ordered = len(customer_orders)
Percentage_of_Products_Ordered = Total_Products_Ordered/len(products) * 100

# Storing the value of customer order statstics to a tuple

order_stat = []
order_stat.append(Total_Products_Ordered)
order_stat.append(Percentage_of_Products_Ordered)
order_stat

order_status = tuple(order_stat)

print()
print()
# Displaying the order statstics
print("Order Statistics: ")
print("Total Products Ordered:", Total_Products_Ordered)
print("Percentage of Products Ordered:", Percentage_of_Products_Ordered,"%")

print()
print()
# updating the inventory dictionary
input_tsh = int(input("Please enter the quantity for the 't-shirt': "))
input_mug = int(input("Please enter the quantity for the 'mug': "))
input_hat = int(input("Please enetr the quantity for the 'hat': "))
input_book = int(input("Please enter the quantity for the 'book': "))
input_keych = int(input("Please enter the quantity for the 'keychain': "))

inventory['t-shirt'] = input_tsh -1 
inventory['mug'] = input_mug -1 
inventory['hat'] = input_hat -1
inventory['book'] = input_book -1
inventory['keychain'] = input_keych -1

print()
# DIsplay only the quantity of each ordered product
print("The quantity of each product is: ", inventory.values())
````
